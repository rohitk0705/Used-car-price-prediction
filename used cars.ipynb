```python
data = pd.read_csv(r"/kaggle/input/used-car-dataset/used cars lecture 4.csv")
print(data) 
```

```python
data.describe()
```

```python
data.info()
```

```python
# calculating mean of columns with NaN
mean_KMS = data.KMS.mean()
mean_mil = data.mileage.mean()
mean_torque = data.torque.mean()
print("mean KMS is",mean_KMS)
print("mean mileage is",mean_mil)
print("mean torque is",mean_torque)
```

```python
data.KMS.fillna(mean_KMS,inplace=True)
data.mileage.fillna(mean_mil,inplace=True)
data.torque.fillna(mean_torque,inplace=True)
```

```python
data.describe()
```

```python
data.info()
```

```python
import seaborn as sns
sns.pairplot(data,height=2,aspect=1)
```

```python
sns.heatmap(data.corr(),annot=True,linewidth=0.6)
```

```python
X=data.drop(["price"], axis=1)
Y=data.price
print (X)
print (Y)
```

```python
# traing and testing data
from sklearn.model_selection import train_test_split
xTrain, xTest, yTrain, yTest=train_test_split(X,Y, test_size=0.2,random_state=42)
```

```python
#Step 3: Modeling/fitting
from sklearn. linear_model import LinearRegression
eqn = LinearRegression()
eqn.fit(xTrain, yTrain)
```

```python
print("the value of intercept is",eqn.intercept_)
print("the values of slopes are",eqn.coef_)
```

```python
yPredict = eqn.predict(xTest)
print(yPredict)
```

```python
from sklearn import metrics
MAE = metrics.mean_absolute_error(yTest,yPredict)
print("the value of MAE is", MAE)
R2 = metrics.r2_score(yTest,yPredict)
print("R2 is", R2)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.regplot(x=yPredict,y=yTest)
```
